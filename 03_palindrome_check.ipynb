# Palindrome Checker

Check if a word or sentence reads the same forwards and backwards.

## Palindrome Function

```python
def is_palindrome(s):
    s = s.lower().replace(' ', '')
    return s == s[::-1]

# Test single words
words = ['Madam', 'racecar', 'hello', 'level', 'Python']
for w in words:
    result = 'Palindrome ✓' if is_palindrome(w) else 'Not a palindrome ✗'
    print(f'{w:12s} → {result}')
```

## Sentence Palindrome

```python
sentences = ['A man a plan a canal Panama', 'Was it a car or a cat I saw', 'Hello World']
for s in sentences:
    print(f'"{s}" → {"Palindrome ✓" if is_palindrome(s) else "Not a palindrome ✗"}')
```
