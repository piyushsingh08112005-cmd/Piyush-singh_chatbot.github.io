# Anagram Checker

Determine if two strings are anagrams of each other.

## Anagram Using Sorted

```python
def is_anagram(a, b):
    return sorted(a.lower()) == sorted(b.lower())

pairs = [('listen', 'silent'), ('hello', 'world'), ('triangle', 'integral'), ('Python', 'nohtyP')]
for a, b in pairs:
    print(f'({a}, {b}) → {"Anagram ✓" if is_anagram(a, b) else "Not an anagram ✗"}')
```

## Using Counter (more Pythonic)

```python
from collections import Counter

def is_anagram_counter(a, b):
    return Counter(a.lower()) == Counter(b.lower())

print(is_anagram_counter('listen', 'silent'))  # True
print(is_anagram_counter('hello', 'world'))    # False
```
