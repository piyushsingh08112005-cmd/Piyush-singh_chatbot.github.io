# Advanced List Comprehensions

Nested comprehensions, conditional filtering, and matrix operations.

## Even Squares

```python
numbers = range(1, 21)
squares_of_evens = [n**2 for n in numbers if n % 2 == 0]
print('Even squares:', squares_of_evens)
```

## Matrix Flatten

```python
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat = [x for row in matrix for x in row]
print('Matrix:', matrix)
print('Flattened:', flat)
```

## Dictionary Comprehension

```python
words = ['apple', 'banana', 'cherry', 'date']
word_lengths = {w: len(w) for w in words}
print('Word lengths:', word_lengths)

# Filter only long words
long_words = {k: v for k, v in word_lengths.items() if v > 5}
print('Long words:', long_words)
```
