# Pandas Data Analysis

Creating DataFrames, filtering, grouping, and summarizing data with Pandas.

## Setup
Install with: `pip install pandas`

```python
import pandas as pd
print('Pandas version:', pd.__version__)
```

## Creating a DataFrame

```python
data = {
    'Name':   ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank'],
    'Age':    [25, 30, 35, 28, 22, 31],
    'Score':  [85, 90, 78, 92, 88, 74],
    'Grade':  ['B', 'A', 'C', 'A', 'B', 'C']
}

df = pd.DataFrame(data)
print(df)
```

## Basic Info & Stats

```python
print('Shape:', df.shape)
print('\nColumn types:\n', df.dtypes)
print('\nDescriptive Stats:\n', df.describe())
```

## Filtering & Selecting

```python
# High scorers
top = df[df['Score'] >= 88][['Name', 'Score', 'Grade']]
print('Top scorers:\n', top)

# Grade A students
grade_a = df[df['Grade'] == 'A']
print('\nGrade A:\n', grade_a)
```

## GroupBy & Aggregation

```python
avg_by_grade = df.groupby('Grade')['Score'].mean()
print('Average score by grade:\n', avg_by_grade)

count_by_grade = df.groupby('Grade').size()
print('\nCount by grade:\n', count_by_grade)
```

## Adding & Modifying Columns

```python
df['Pass'] = df['Score'] >= 80
df['Score_Normalized'] = (df['Score'] - df['Score'].min()) / (df['Score'].max() - df['Score'].min())
df['Score_Normalized'] = df['Score_Normalized'].round(2)
print(df)
```

## Exporting to CSV

```python
df.to_csv('students.csv', index=False)
print('Saved to students.csv')

# Read it back
df2 = pd.read_csv('students.csv')
print(df2.head())
```
